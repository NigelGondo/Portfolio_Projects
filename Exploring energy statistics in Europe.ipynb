# Project summary

This project aimed to conduct exploratory data analysis of the energy statistics in Europe that touched on supply, consumption, and prices of energy. The period of the project covered data from 2011 to 2022 and focused on 40 countries. There are several types of energy included in this analysis and a few of them include fossil fuels, gas, and electricity, and the unit of measure for all these energy types is “thousand tonnes of oil equivalent”. This unit of measure is the amount of energy released by burning one tonne of crude oil and it is approximately 42 gigajoules or 11.630 megawatt-hours.

Also, the project touched on the relationship that energy has with the GDP of a country and showed why energy is significant for any country's economy.

# Importing and cleaning data

## Energy supply data 

```python
#importing energy supply data using eurostat api
df_energy_supply = eurostat.get_data_df('ten00122')
df_energy_supply.head()
```

**NOTE:** 

The column **siec** stands for **standard international energy product classification** and this basically classifies the different types of energy supplied and consumed. The types of energy that fall under siec are:
- Solid fossil fuels
- Manufactured gases
- Electricity
- Natural gas
- Heat
- Nuclear heat,
- Oil and petroleum (excluding biofuel portion)
- Peat and peat products
- Renewables and biofuels
- Oil shale and oil sands
- Non-renewable waste


The column geo/TIME_PERIOD is geographical data with country codes and the value of these country codes will be changed to the full country name when cleaning the dataframe. 

```python
#checking number of unique countries
df_energy_supply['geo\TIME_PERIOD'].unique()
```

```python
#checking number of unique types of energy
df_energy_supply['siec'].unique()
```

```python
#REPLACING VALUES IN THE geo/TIME_PERIOD AND siec COLUMNS TO THE FULL DESCRIPTION OF THE VALUES

#creating dictionary for code and country name
CountryCode_CountryName = {'AL':'Albania', 'AT':'Austria', 'BA':'Bosnia and Herzegovina', 
                           'BE':'Belgium', 'BG':'Bulgaria', 'CY':'Cyprus', 
                           'CZ':'Czechia', 'DE':'Germany', 'DK':'Denmark', 
                           'EE':'Estonia', 'EL':'Greece','ES':'Spain', 
                           'EU27_2020':'European Union-27 countries (from 2020)', 
                           'FI':'Finland', 'FR':'France', 
                           'GE':'Georgia', 'HR':'Croatia', 'HU':'Hungary', 
                           'IE':'Ireland', 'IS':'Iceland', 'IT':'Italy',
                           'LT':'Lithuania', 'LU':'Luxembourg', 'LV':'Latvia', 
                           'MD':'Moldova', 'ME':'Montenegro', 'MK':'North Macedonia', 
                           'MT':'Malta', 'NL':'Netherlands', 'NO':'Norway', 
                           'PL':'Poland', 'PT':'Portugal','RO':'Romania', 
                           'RS':'Serbia', 'SE':'Sweden', 'SI':'Slovenia', 
                           'SK':'Slovak Republic', 'TR':'Turkiye', 'UA':'Ukraine', 
                           'UK':'United Kingdom', 'XK':'Kosovo'}

#creating dictionarty for energy code and energy name
Type_of_energy = {'C0000X0350-0370':'Solid fossil fuels', 'C0350-0370':'Manufactured gases', 
                  'E7000':'Electricity', 'G3000':'Natural gas', 
                  'H8000':'Heat','N900H':'Nuclear heat', 
                  'O4000XBIO':'Oil and petroleum (excluding biofuel portion)', 
                  'P1000':'Peat and peat products', 
                  'RA000':'Renewables and biofuels', 'S2000':'Oil shale and oil sands', 
                  'TOTAL':'Total','W6100_6220':'Non-renewable waste'}

#replacing values
df_energy_supply['geo\TIME_PERIOD'] = df_energy_supply['geo\TIME_PERIOD'].replace(CountryCode_CountryName)

df_energy_supply['siec'] = df_energy_supply['siec'].replace(Type_of_energy)
```

```python
#checking that country code values have been changed to country name
df_energy_supply['geo\TIME_PERIOD'].unique()
```

```python
#checking that energy product classification code has been changed to the name of the energy product
df_energy_supply['siec'].unique()
```

```python
#UNPIVOTING THE DATAFRAME SO THAT THERE IS ONE COLUMN FOR YEAR AND ONE COLUMN FOR THE VALUES

#dropping unnecessary columns and unpivoting dataframe
df_energy_supply.drop(columns=['freq', 'nrg_bal', 'unit'], 
                      axis=1, 
                      inplace=True)

#creating list Of the column names to be unpivoted
energy_supply_value_columns = list(df_energy_supply.columns[2:])

#unpivoting the year columns 
df_energy_supply_unpivoted = pd.melt(df_energy_supply, 
                                     id_vars=['siec', 'geo\TIME_PERIOD'], 
                                     value_vars=energy_supply_value_columns, 
                                     value_name='Energy supply: Thousand tonnes of oil equivalent', 
                                     var_name='Year')
#renaming columns
df_energy_supply_unpivoted\
.rename(columns={'siec':'Standard international energy product classification', 
                 'geo\TIME_PERIOD':'Country'}, 
        inplace=True)

df_energy_supply_unpivoted.head()
```

```python
#dropping the Total values
df_energy_supply_unpivoted\
.drop(df_energy_supply_unpivoted\
      [df_energy_supply_unpivoted['Standard international energy product classification']=='Total']\
      .index, 
      inplace=True)

#dropping the European Union-27 countries (from 2020) values
df_energy_supply_unpivoted\
.drop(df_energy_supply_unpivoted\
      [df_energy_supply_unpivoted['Country']=='European Union-27 countries (from 2020)']\
      .index, 
      inplace=True)
```

```python
#checking for null values
df_energy_supply_unpivoted.isnull().sum()
```

**NOTE:** 

There are instances that some countries have missing data, and this data will be filled in by the median based on the country and the energy product. The median is a better measure than the mean due to the possibility of outliers present in the data.

```python
#filling in missing values with the median based on the country and energy supply
df_energy_supply_unpivoted['Energy supply: Thousand tonnes of oil equivalent']=\
df_energy_supply_unpivoted['Energy supply: Thousand tonnes of oil equivalent']\
.fillna(df_energy_supply_unpivoted\
        .groupby(['Standard international energy product classification', 
                  'Country'])['Energy supply: Thousand tonnes of oil equivalent']\
        .transform('median')) 
```

```python
#checking that null values have been replaced with the median
df_energy_supply_unpivoted.isnull().sum()
```
