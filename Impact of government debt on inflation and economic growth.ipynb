# Brief description of project

**Project Summary: Investigating the Impact of Government Gross Debt on Inflation and Economic Growth**

In this research endeavor, I delved into the intricate relationship between government gross debt, inflation, and economic across both advanced and developing economies. The study encompassed several crucial stages:

1. **Data Wrangling**:
   - I meticulously handled data by performing tasks such as data import, removal of unnecessary rows, column unpivoting, and imputing missing values.

2. **Exploratory Data Analysis (EDA)**:
   - I conducted descriptive statistical analyses on relevant variables.
   - Time series analysis was employed to understand the temporal trends of these variables.
   - Boxplots were used to visualize the distributions of key economic indicators.
   - Aggregated data facilitated the creation of bar charts, allowing us to explore the top 10 and bottom 10 countries in terms of debt, inflation, and GDP.
   - Scatter plots visually depicted the relationship between gross debt and both inflation and GDP growth.

3. **Hypothesis Testing**:
   - To validate my hypotheses, I applied the **Chi-Square test of independence**.

Overall, this investigation sheds light on the complex interplay between economic indicators, offering valuable insights for policymakers and researchers.

## Advanced economies

```python
#importing advanced economies data
advanced_economies = pd.read_csv("Advanced_economies_data.csv")
advanced_economies.head()
```

```python
#dropping unnecessary columns
advanced_economies.drop(columns=['Units',
                                 'Scale',
                                 'Country/Series-specific Notes'], 
                        axis=1, 
                        inplace=True)
```

```python
#checking that columns are dropped
advanced_economies.head()
```

```python
#unpivoting the columns of years so as to create a year column and values column

#creating list of columns to be unpivoted
value_columns = list(advanced_economies.columns[5:-1])

#unpivoting the year columns so that one column is for the year and the other column for values
advanced_economies_unpivot = pd.melt(advanced_economies, 
                                     id_vars=['Economy type', 
                                              'Country', 
                                              'Continent', 
                                              'Region',
                                              'Subject Descriptor'],
                                     value_vars=value_columns, 
                                     var_name='Year',
                                     value_name='value')

advanced_economies_unpivot.head()
```

```python
#pivoting the subject descriptor column so that each variable has its own column
advanced_economies_pivot = \
advanced_economies_unpivot\
.pivot(index=['Economy type',
              'Country', 
              'Continent', 
              'Region',
              'Year'], 
       columns='Subject Descriptor', 
       values='value')\
.reset_index()\
.rename(columns={'General government gross debt':'General government gross debt (Percentage of GDP)',
                 'General government net debt':'General government net debt (Percentage of GDP)',
                 'Gross domestic product, current prices':'Gross domestic product, current prices (Nominal GDP-USD Billions)',
                 'Inflation, average consumer prices':'Inflation, average consumer prices (Percentage change)'})

advanced_economies_pivot.head()
```

```python
#checking data types
advanced_economies_pivot.dtypes
```

```python
#changing the year data type to int
advanced_economies_pivot['Year'] = advanced_economies_pivot['Year'].astype(int)
                                                            
advanced_economies_pivot.dtypes
```

In the dataset, several missing values are observed. Therefore, we will conduct an analysis to determine the percentage of missing values for each country.

```python
#creating a dataframe that summaries the percentage of missing values of the variables to be analysed

#creating list of countries and empty list
country_list_1 = list(advanced_economies_pivot['Country'].unique())
series_list_1 = []
n = 10

#creating a loop to iterate through each country creating a series of the summarised\
#missing data then creating a dataframe
for i in country_list_1:
    country_filter_1 = list(advanced_economies_pivot\
                        .loc[(advanced_economies_pivot['Country']==i)]\
                        .isnull()\
                        .sum() / len(advanced_economies_pivot\
                                     .loc[(advanced_economies_pivot['Country']==i)])*100)
    for j in country_filter_1:
        series_list_1.append(j)
        sublists_1 = [series_list_1[i:i+n] for i in range(0,len(series_list_1),n)]
        df_1 = pd.DataFrame(sublists_1)
```

Several countries exhibit substantial missing data, particularly concerning the variable of net debt. Within the dataset, numerous countries lack data for the initial 20 years spanning from 1980 to 2000. To simplify the analysis, we will narrow the time frame, focusing on data from the year 2000 to 2023. This adjustment will significantly reduce the number of missing values, and any remaining gaps can be more easily addressed by filling in with the median.

```python
#filtering the data from 2000 to 2023
advanced_economies_pivot_2000to2023 = advanced_economies_pivot[advanced_economies_pivot['Year']\
                                                               .between(2000,2023)]
advanced_economies_pivot_2000to2023.head()
```

```python
#using a similar loop as before to calculate percentage of missing values by country

#creating list of countries and empty list
country_list_2 = list(advanced_economies_pivot_2000to2023['Country'].unique())
series_list_2 = []
n = 10

#creating a loop to iterate through each country creating a series of the summarised\
#missing data then creating a dataframe
for i in country_list_2:
    country_filter_2 = list(advanced_economies_pivot_2000to2023\
                        .loc[(advanced_economies_pivot_2000to2023['Country']==i)]\
                        .isnull()\
                        .sum() / len(advanced_economies_pivot_2000to2023\
                                     .loc[(advanced_economies_pivot_2000to2023['Country']==i)])*100)
    for j in country_filter_2:
        series_list_2.append(j)
        sublists_2 = [series_list_2[i:i+n] for i in range(0,len(series_list_2),n)]
        df_2 = pd.DataFrame(sublists_2)
```

```python
df_2
```

```python
#renaming the index and the columns
columns_2 = list(advanced_economies_pivot_2000to2023\
             .loc[(advanced_economies_pivot_2000to2023['Country']=='Andorra')]\
             .isnull()\
             .sum()\
             .index)
df_2.index = country_list_2
df_2.columns = columns_2

df_2
```

Handling extensive missing data can be challenging. To mitigate the risk of data distortion caused by imputing missing values, we will focus exclusively on countries with less than 5 percent missing data.

```python
#Checking the countries that have more than 5 percent of missing data
list_of_countries_to_drop =\
list(df_2[(df_2['General government gross debt (Percentage of GDP)']>5) |\
     (df_2['Gross domestic product, current prices (Nominal GDP-USD Billions)']>5) |\
     (df_2['Gross domestic product, deflator']>5) |\
     (df_2['Inflation, average consumer prices (Percentage change)']>5)].index)

list_of_countries_to_drop
```

```python
#dropping the countries that have more than 5 percent of missing data
advanced_economies_pivot_2000to2023_final = advanced_economies_pivot_2000to2023\
[~advanced_economies_pivot_2000to2023\
 ['Country']\
 .isin(list_of_countries_to_drop)]
```

```python
#checking number of advanced economies left
advanced_economies_pivot_2000to2023_final['Country'].nunique()
```

```python
#list of the remaining advanced economies
advanced_economies_pivot_2000to2023_final['Country'].unique()
```

There are now 36 countries left in the dataframe.

```python
#checking there are no duplicate entries in the dataset
advanced_economies_pivot_2000to2023_final.duplicated().sum()
```

No duplucate entries in the dataset

```python
#dropping government net debt column
advanced_economies_pivot_2000to2023_final.drop(columns='General government net debt (Percentage of GDP)', 
                                           inplace=True)
```

```python
# Only 1 value of gross debt needs to be filled in and will use the median for this
advanced_economies_pivot_2000to2023_final.isnull().sum()
```

```python
#filling in the variables that have missing values with the median based on the country

#filling missing values of government gross debt with median
advanced_economies_pivot_2000to2023_final['General government gross debt (Percentage of GDP)'] =\
advanced_economies_pivot_2000to2023_final['General government gross debt (Percentage of GDP)']\
.fillna(advanced_economies_pivot_2000to2023_final\
        .groupby('Country')['General government gross debt (Percentage of GDP)']\
        .transform('median'))
```

```python
#checking that missing values have been filled
advanced_economies_pivot_2000to2023_final.isnull().sum()
```

```python
#creating new columns for further analysis

#calculating real GDP  by dividing the nominal GDP with the GDP deflator
advanced_economies_pivot_2000to2023_final['Gross domestic product, constant prices (Real GDP-USD Billions)'] = \
advanced_economies_pivot_2000to2023_final['Gross domestic product, current prices (Nominal GDP-USD Billions)'] / \
(advanced_economies_pivot_2000to2023_final['Gross domestic product, deflator'] / 100)

#calulation actual government gross debt by multipling the percentage with nominal gdp
advanced_economies_pivot_2000to2023_final['General government gross debt (USD-Billions)'] =\
(advanced_economies_pivot_2000to2023_final['General government gross debt (Percentage of GDP)'] / 100) * \
advanced_economies_pivot_2000to2023_final['Gross domestic product, current prices (Nominal GDP-USD Billions)']

#calculating GDP growth of nominal GDP
advanced_economies_pivot_2000to2023_final['GDP growth (percentage change Nominal GDP)'] =\
advanced_economies_pivot_2000to2023_final\
.groupby('Country')['Gross domestic product, current prices (Nominal GDP-USD Billions)']\
.pct_change() * 100

#calculating GDP growth of real GDP
advanced_economies_pivot_2000to2023_final['GDP growth (percentage change Real GDP)'] =\
advanced_economies_pivot_2000to2023_final\
.groupby('Country')['Gross domestic product, constant prices (Real GDP-USD Billions)']\
.pct_change() * 100

#filling in the year 2000 growth with the median
advanced_economies_pivot_2000to2023_final['GDP growth (percentage change Nominal GDP)'] =\
advanced_economies_pivot_2000to2023_final['GDP growth (percentage change Nominal GDP)']\
                                                .fillna(advanced_economies_pivot_2000to2023_final\
                                                .groupby('Country')['GDP growth (percentage change Nominal GDP)']\
                                                .transform('median'))

#filling in the year 2000 growth with the median
advanced_economies_pivot_2000to2023_final['GDP growth (percentage change Real GDP)'] =\
advanced_economies_pivot_2000to2023_final['GDP growth (percentage change Real GDP)']\
                                                .fillna(advanced_economies_pivot_2000to2023_final\
                                                .groupby('Country')['GDP growth (percentage change Real GDP)']\
                                                .transform('median'))

advanced_economies_pivot_2000to2023_final.head()
```

## Distributions

I will employ a boxplot for the examination of the distribution patterns of key variables such as gross debt as a percentage of GDP, inflation rate, and GDP growth. This will also aid in the identification of outliers.

```python
#creating a function to create subplots of boxplots
def box_plots(Economy_type, Continent, Column_name, Title_1, Title_2, Title_3, Title_4, Main_title):
    
    #Splitting advanced economies and developing economies
    advance_filter = df_advanced_developing_economies_data\
    .loc[(df_advanced_developing_economies_data['Economy type']=='Advanced economies')]
    
    developing_filter = df_advanced_developing_economies_data\
    .loc[(df_advanced_developing_economies_data['Economy type']=='Emerging markets and developing economies')]
    
    fig, ax = plt.subplots(2,2, figsize=(20,23))
    
    #creating first boxplot
    
    #calculating medians then plotting them on the boxplot
    medians_1 = advance_filter.groupby([Economy_type])[Column_name].median().round(2)
    vertical_offset_1 = advance_filter[Column_name].median() * 0.05 # offset from median for display
    
    boxplot_1 = sns.boxplot(data=advance_filter,
                            y=Column_name, 
                            ax=ax[0,0])
    
    for xtick in boxplot_1.get_xticks():
        boxplot_1.text(xtick,medians_1[xtick] + vertical_offset_1,medians_1[xtick], 
                horizontalalignment='center',size='large',color='black',weight='semibold')
    #setting title and adjusting tick labels
    ax[0,0].set_title(Title_1, 
                      fontweight='bold', 
                      fontsize=20)    
    ax[0,0].set_xticklabels(list(medians_1.index),
                            fontsize=13)
    ax[0,0].tick_params(axis='y', 
                        labelsize=13)
     
    #creating second boxplot
    
    #calculating medians then plotting them on the boxplot
    medians_2 = advance_filter.groupby([Continent])[Column_name].median().round(2).sort_values(ascending=True)
    vertical_offset_2 = advance_filter[Column_name].median() * 0.05 # offset from median for display
    
    boxplot_2 = sns.boxplot(data=advance_filter, 
                            x=Continent, 
                            y=Column_name, 
                            order=medians_2.index,
                            ax=ax[0,1])
    for xtick in boxplot_2.get_xticks():
        boxplot_2.text(xtick,medians_2[xtick] + vertical_offset_2,medians_2[xtick], 
                horizontalalignment='center',size='large',color='black',weight='semibold')
    #setting title and adjusting tick labels
    ax[0,1].set_title(Title_2, 
                      fontweight='bold', 
                      fontsize=20) 
    ax[1,1].set_xticklabels(list(medians_2.index),
                            fontsize=13)
    ax[0,1].tick_params(axis='y', 
                        labelsize=13)

    #creating third boxplot
    #calculating medians then plotting them on the plot
    medians_3 = developing_filter.groupby([Economy_type])[Column_name].median().round(2)
    vertical_offset_3 = developing_filter[Column_name].median() * 0.05 # offset from median for display

    boxplot_3 = sns.boxplot(data=developing_filter,
                            y=Column_name, 
                            ax=ax[1,0])
    
    for xtick in boxplot_3.get_xticks():
        boxplot_3.text(xtick,medians_3[xtick] + vertical_offset_3,medians_3[xtick], 
                horizontalalignment='center',size='large',color='black',weight='semibold')
    #setting title and adjusting tick labels
    ax[1,0].set_title(Title_3, 
                      fontweight='bold',
                      fontsize=20)    
    ax[1,0].set_xticklabels(list(medians_3.index),
                            fontsize=13)
    ax[1,0].tick_params(axis='y', 
                        labelsize=13)
     
    #creating fourth boxplot 
     #calculating medians then plotting them on the plot
    medians_4 = developing_filter.groupby([Continent])[Column_name].median().round(2).sort_values(ascending=True)
    vertical_offset_4 = developing_filter[Column_name].median() * 0.05 # offset from median for display

    boxplot_4 = sns.boxplot(data=developing_filter, 
                            x=Continent, 
                            y=Column_name, 
                            order=medians_4.index,
                            ax=ax[1,1])
   
    for xtick in boxplot_4.get_xticks():
        boxplot_4.text(xtick,medians_4[xtick] + vertical_offset_4,medians_4[xtick], 
                horizontalalignment='center',size='large',color='black',weight='semibold')
    #setting title and adjusting tick labels
    ax[1,1].set_title(Title_4, 
                      fontweight='bold', 
                      fontsize=20) 
    ax[1,1].set_xticklabels(list(medians_4.index),
                            fontsize=13)
    ax[1,1].tick_params(axis='y', 
                        labelsize=13)
    
    #main title
    plt.suptitle(Main_title, 
                 fontweight='bold', 
                 fontsize=25, 
                 y=1.05)
    
    plt.tight_layout()
    
#calling function for plotting boxplots  
box_plots('Economy type', 
          'Continent', 
          'General government gross debt (Percentage of GDP)', 
          'Overall distribution of advanced economies', 
          'Distribution of advanced economies by continent', 
          'Overall distribution of emerging markets and developing economies', 
          'Distribution of emerging markets and developing economies by continent',
          'Distribution of government gross debt (Pecentage of GDP)')
```

# Hypothesis testing: Chi-Square test of independence

The **chi-square test of independence** assesses whether there exists a significant association or dependence between two categorical variables. Specifically:

1. **Purpose and Data Type**:
   - It is employed for nonparametric data, which means data that do not adhere to a specific distribution.
   - The test aims to determine whether the two categorical variables are related in the population.

2. **Contingency Table**:
   - The test evaluates whether the observed frequencies of categories in a contingency table (also known as a cross-tabulation) significantly deviate from the expected frequencies if the variables were unrelated.
   - The contingency table displays the number of observations in each combination of groups.

3. **Test Statistic and Interpretation**:
   - The chi-square test calculates a test statistic based on the observed frequencies in the contingency table.
   - Expected frequencies are then compared to the observed frequencies. When the variables are unrelated, the observed and expected frequencies tend to be similar.
   - If the test statistic is statistically significant, it suggests that the variables are associated or related.

In summary, the chi-square test of independence provides insights into the relationships between categorical variables, making it valuable for decision-making and research.

## Advanced economies test 

### Association of debt and inflation

The government gross debt, inflation rate, and GDP growth variables will be categorized in preparation for conducting the Chi-Square test.

- We will categorize the government gross debt variable into two groups using the median. The first category, termed moderate, will include values below the median, while the second category will encompass values above the median. For advanced economies, the median stands at approximately 57 percent. Consequently, the moderate range will span from **0 to 57 percent**, while the high category will cover values from **58 to 261 percent**, which represents the highest government gross debt value in the dataframe.

- Inflation will be classified into two groups: **moderate** and **high** inflation. The moderate category will span from **-2% to 4%**, while the high category will encompass values **above 4% up to 10%**.

- GDP growth will be classified into three groups of low, medium and high. Low growth will span from **-4% to 1%**, medium growth will be above **1% to 3%** and high will encompass values **above 3% up to 10%**.





```python
#creating categories of gross debt, inflation rate and gdp growth

#categorising government gross debt
advanced_economies_pivot_2000to2023_final['Debt category']= \
pd.cut(x=advanced_economies_pivot_2000to2023_final['General government gross debt (Percentage of GDP)'], 
       bins=[0, 57, 262], 
                     labels=['Moderate', 'High'])

#categorising inflation
advanced_economies_pivot_2000to2023_final['Inflation category']= \
pd.cut(x=advanced_economies_pivot_2000to2023_final['Inflation, average consumer prices (Percentage change)'], 
       bins=[-3, 4, 10], 
                     labels=['Moderate','High'])

#categorising gdp growth
advanced_economies_pivot_2000to2023_final['Real GDP growth category']= \
pd.cut(x=advanced_economies_pivot_2000to2023_final['GDP growth (percentage change Real GDP)'], 
       bins=[-5, 1, 3, 10], labels=['Low', 'Medium', 'High'])
```

```python
#frequency summary
rp.summary_cat(advanced_economies_pivot_2000to2023_final[['Debt category', 
                                                          'Inflation category']])
```

#### Stating the null and alternative hypothesis

$H_{0}$: There is no association between government gross debt and inflation rate

$H_{1}$: There is an association between government gross debt and inflation rate


#### level of significance

```python
#5% level of significance 
alpha__debt_inf_adv = 0.05
```

#### Computing the value of the test statistic and the p-value

```python
crosstab, test_results, expected = \
rp.crosstab(advanced_economies_pivot_2000to2023_final['Debt category'], 
            advanced_economies_pivot_2000to2023_final['Inflation category'],
                                               test= 'chi-square',
                                               expected_freqs= True,
                                               prop= 'cell')

crosstab
```

```python
test_results
```

#### Decision

```python
alpha__debt_inf_adv < test_results.results[1]
```

The **p-value** is less than the **5% significance level** (0.05), leading us to **reject the null hypothesis**. Consequently, the test results are **statistically significant at the 5% level**, providing **strong evidence** that **government gross debt affects the inflation rate of advanced economies**.
