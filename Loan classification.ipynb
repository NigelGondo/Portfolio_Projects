# Brief description


This project is about bank loans, and based on the personal details of clients, banks will have the option to either approve or reject a client's loan application. Therefore, the main objective of this project was to create a binary classification model that aimed to classify the status of loan applications as approved or rejected based on the client's details which included gender, marital status, applicant income, co-applicant income, dependents, employment type, loan amount, terms, credit history, and area of residents of the applicant. I ran several classification algorithms, and the algorithm that produced the best prediction was **Support Vector Machine (SVM)**. The data I sourced from Kaggle repository.

**A brief note on SVM**

SVM is used for both classification and regression problems. The objective of this algorithm is to create a decision boundary that separates n-dimensional space into classes so that any new data point can be accurately placed in its class. The best decision boundary is called a hyperplane.

```python
#columns that have null values fillin them up with the mode frequent value in that columns

#creating a copy of the loan data
df_loan_nulls_filled = df_loan.copy()

#creating a list of the columns with missing values
columns_to_be_filled = ['Gender', 'Married', 'Dependents', 'Self_Employed', 'Term', 'Credit_History']

#creating a for loop to fill the columns with missing values with the frequent value in that column
for i in columns_to_be_filled:
    df_loan_nulls_filled[i].fillna(df_loan_nulls_filled[i].mode()[0], inplace=True)
```

### Removing outliers using the interquatile range method

```python
#Removing outliers using the interquatile range
def outliers(df, feature):
    Q1= df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR
    return upper_limit, lower_limit

upper, lower = outliers(df_loan_nulls_filled, 'Applicant_Income')
print("Upper whisker: ", upper)
print("Lower Whisker: ", lower)
```

```python
#Viewing the applicant income outliers
df_loan_nulls_filled[(df_loan_nulls_filled['Applicant_Income'] < lower) | (df_loan_nulls_filled['Applicant_Income'] > upper)]
```

```python
#Creating new dataframe without outliers
df_loan_no_outliers_1 = df_loan_nulls_filled[(df_loan_nulls_filled['Applicant_Income'] > lower) & \
                                        (df_loan_nulls_filled['Applicant_Income'] < upper)]
print('Number of records before outliers are removed:' + str(len(df_loan_nulls_filled)))
print('Number of records after outliers are removed:' + str(len(df_loan_no_outliers_1))) 
#the two outliers were removed
```

# Predictive modelling without outliers included

## Encoding and standardising the variables

```python
#independent variables to be one hot encoded
df_loan_train_encode_no_outliers = df_loan_no_outliers_3[['Gender', 
                                                          'Married', 
                                                          'Dependents',
                                                          'Education', 
                                                          'Self_Employed', 
                                                          'Area']]

#independent variables to be standardised
df_loan_train_standardise_no_outliers = df_loan_no_outliers_3[['Applicant_Income', 
                                                               'Coapplicant_Income', 
                                                               'Loan_Amount', 
                                                               'Term']]

#dependent variables to be label encoded
df_loan_train_categorise_no_outliers = df_loan_no_outliers_3['Status']
```

```python
#One hot encoding 
ct = ColumnTransformer(transformers=[('encoder',
                                        OneHotEncoder(), 
                                        [0,1,2,3,4,5])], 
                       remainder='passthrough')

df_loan_train_encode_no_outliers_numpy = np.array(ct.fit_transform(df_loan_train_encode_no_outliers))

df_loan_train_encode_no_outliers_numpy[:5]
```

**Note:**

For those not familiar with one hot encoding, this is a method of converting each categorical value into a new categorical column and assign a binary value of 1 or 0 to those columns. Each integer value is represented as a binary vector. All the values are zero, and the index is marked with a 1.

The benefit with one hot encoding is that data will have no relationship with each other reducing any distortion on the models performance.

```python
#standarding
sc = StandardScaler()
df_loan_train_standardise_no_outliers_numpy = sc.fit_transform(df_loan_train_standardise_no_outliers)
df_loan_train_standardise_no_outliers_numpy[:5]
```

```python
#label encoding the dependent variable
le = LabelEncoder()
df_loan_train_categorise_no_outliers_numpy = le.fit_transform(df_loan_train_categorise_no_outliers)
df_loan_train_categorise_no_outliers_numpy[0:5]
```

## Train-Test data and fitting the model

```python
#converting the credit history column to numpy array
df_loan_credit_his_numpy = df_loan_no_outliers_3.Credit_History.to_numpy()
df_loan_credit_his_numpy = df_loan_credit_his_numpy.reshape(len(df_loan_credit_his_numpy),1)

#concatenating the encoded and standardised arrays to create the independent array
X = np.concatenate((df_loan_train_encode_no_outliers_numpy, 
                    df_loan_train_standardise_no_outliers_numpy,
                    df_loan_credit_his_numpy), 
                   axis = 1)

#dependent variable array
y = df_loan_train_categorise_no_outliers_numpy
```

```python
#train-test split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 40)
```

```python
#fitting the model
model_svc = SVC()

model_svc.fit(x_train,y_train)
```

## Model evaluation

### Classification report

```python
#predicting y
y_pred_svc = model_svc.predict(x_test)

#accuracy and classification report
print('Accuracy score:', str(round((accuracy_score(y_test,y_pred_svc))*100,2)) + str('%'))
print(classification_report(y_test, y_pred_svc, target_names=['No','Yes']))
```

**COMMENTS ON THE CLASSIFICATION REPORT**

In the classification report there are 3 important metrics which are the precision, recall and f1-score:

-> Precision measures the percentage of correct true positive predictions relative to the total of true and false positives: For the model the precision is 82%.

-> Recall measures the percentage of correct true positive predictions relative to total of true positives and false negatives: For the model the recall is 99%.

-> F1-score is a weighted mean of precision and recall. The closer the value of F1 is to 1, the better the expected performance: for the model the F1-score is 90%.

-> Support is the number of actual occurances of the class in the dataset.

### Confusion matrix

```python
#plotting confusion matrix
ConfusionMatrixDisplay.from_estimator(model_svc, x_test, y_test, display_labels=['No','Yes'])
plt.show()
```

**Comments on confusion matrix**

A confusion matrix summarises the prediction results of a classification problem. The benefit of this matrix is that it gives clarity on the types of errors being made by the classifier. It is also used to determine the classifier performance based on true positive (TP), true negative (TN), false positive (FN) and false negative (FN). In the case of this model:

TN -> 11 applicants were classified correctly having their loan applications rejected.

TP -> 73 applicants were classified correctly as having their loan application accepted.

FN -> 16 applicants were misclassified as having their loan application rejected.

FP -> 1 applicant was misclassified as having their loan application accepted.

### k-fold cross validation to test the accuracy of the model (without outliers)

K-fold cross validation is a resampling procedure with the purpose of evaluating an ML model on how well it can predict the outcome of unseen data.

```python
#running the cross validation 10 times
accuracies = cross_val_score(estimator=model_svc,
                             X=x_train, 
                             y=y_train,
                             cv=10)

print("Accuracy: {:.2f}".format(accuracies.mean()*100) + str('%'))

print("Std: {:.2f} ".format(accuracies.std()*100))
```
