# Brief Description

The war between Ukraine and Russia has had a negative domino effect across multiple regions of the world both economically and politically. However, this project aimed to investigate how the war has affected key food commodities that come from Ukraine and Russia. The key commodities produced and exported from Ukraine and Russia are wheat, barley, maize, and sunflower oil.

```python
#importing the relevant libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
import seaborn as sns
%matplotlib inline
sns.set()
sns.set_style('white')
plt.rcParams.update({'figure.figsize':(15,10), 'figure.dpi':100})
import warnings
warnings.filterwarnings('ignore')
```

# Data cleaning and EDA

### Wheat Data

```python
#rearranging the columns to start with entity and year
df_wheat = df_wheat[['Entity',
                     'Year',
                     'Domestic wheat supply',
                     'Wheat exports', 
                     'Wheat imports',
                     'Wheat stocks', 
                     'Net wheat imports',
                     'Wheat imports (% domestic supply)', 
                     'Wheat stocks (% domestic supply)', 
                     'Wheat imports from Ukraine', 
                     'Wheat imports from Russia',
                     'Wheat imports from Ukraine + Russia',
                     'Wheat imports from Ukraine (% imports)',
                     'Wheat imports from Russia (% imports)',
                     'Wheat imports from Ukraine + Russia (% imports)',
                     'Wheat imports from Ukraine (% supply)',
                     'Wheat imports from Russia (% supply)',
                     'Wheat imports from Ukraine + Russia (% supply)']]
```

```python
#In the Entity column there is a mixture of continents and countries so we will create a seperate data frame for countries only
continents_world = ['Africa', 'Asia', 'Europe', 'Oceania', 'South America', 'World']

df_wheat_countries = df_wheat[~df_wheat.Entity.isin(continents_world)]
```

```python
#Renaming the Entity column to Country and Continent for the new data frame
df_wheat_countries = df_wheat_countries.rename(columns={'Entity':'Country'})
```

```python
#filling all NaN values with zero 
df_wheat_countries.fillna(0, inplace =True)
```

```python
df_wheat_countries.to_csv('df_wheat_countries.csv',
                          index = False) #saving final dataframe as csv

df_wheat_countries.head()
```

### Importing global export and production by Ukraine and Russia of it food commodities

```python
global_exports_Ukr_Rus = pd.read_csv('food-exports-ukraine-russia.csv').\
                         sort_values('ukraine_share_global_exports', ascending=False)
global_exports_Ukr_Rus
```

```python
global_production_Ukr_Rus = pd.read_csv('food-production-ukraine-russia.csv').\
                            sort_values('ukraine_share_global_production', ascending=False)
global_production_Ukr_Rus
```

```python
#merging the two data frames
global_exports_production = pd.merge(left = global_exports_Ukr_Rus,
                                    right = global_production_Ukr_Rus,
                                    left_on= 'Entity',
                                    right_on = 'Entity')
global_exports_production
```

# Data Visualisation

### Global exports and production for Ukraine and Russia

```python
fig, ax = plt.subplots(2)

ax[0].bar('Entity', 'ukraine_share_global_exports',
        color = 'mediumblue',
        data = global_exports_production)

ax[0].bar('Entity', 
        'russia_share_global_exports',
        bottom ='ukraine_share_global_exports',
        color = 'darkgrey',
        data = global_exports_production)

ax[0].set_title('Ukraine and Russia global share of exports',  
                fontsize = 18, 
                fontweight = 'bold')

ax[0].set_xlabel('Commodities', 
               fontsize = 12, 
               fontweight = 'bold')
ax[0].set_ylabel('Percentage of global exports', 
               fontsize = 12, 
               fontweight = 'bold')

ax[0].legend(['Ukraine', 'Russia'])

ax[0].set_frame_on(False)


ax[1].bar('Entity', 'ukraine_share_global_production',
        color = 'mediumblue',
        data = global_exports_production)

ax[1].bar('Entity', 
        'russia_share_global_production',
        bottom ='ukraine_share_global_production',
        color = 'darkgrey',
        data = global_exports_production)

ax[1].set_title('Ukraine and Russia global share of production', 
                fontsize = 18, 
                fontweight = 'bold')

ax[1].set_xlabel('Commodities', 
               fontsize = 12, 
               fontweight = 'bold')
ax[1].set_ylabel('Percentage of global production', 
               fontsize = 12, 
               fontweight = 'bold')

ax[1].legend(['Ukraine', 'Russia'])

ax[1].set_frame_on(False)

plt.suptitle('The share of global production and exports for Ukraine and Russia as of 2019', 
             fontsize = 18, 
             fontweight = 'bold')

plt.tight_layout()

plt.show()
```

From the stacked bar chart above it is evident that Ukraine and Russia produce a large chunk of the global sunflower oil, where the two countries combined to produce over 50 per cent.

In contrast, sunflower oil is the most exported commodity with a combined global export of over 60 per cent followed by wheat with over 20 per cent.

### Time series of commodity prices

```python
#plotting time series for wheat
fig, ax = plt.subplots()

ax.plot(df_wheat_prices.index, 
        'Close', 
        data = df_wheat_prices)

loc = plticker.MultipleLocator(base=20.0)

ax.xaxis.set_major_locator(loc)

plt.xticks(rotation = 45)

plt.title('Time Series of Wheat Prices', 
          fontsize = 18,
          Fontweight = 'bold')

plt.box(False)

plt.show()
```

```python
#plotting time series for maize
fig, ax = plt.subplots()
ax.plot(df_maize_prices.index, 
        'Price', 
        data = df_maize_prices)

locx = plticker.MultipleLocator(base=20.0)

locy = plticker.MultipleLocator(base=20.0)

ax.xaxis.set_major_locator(locx)

ax.yaxis.set_major_locator(locy)
plt.xticks(rotation = 45)

plt.title('Time Series of Maize Prices', 
          fontsize = 18,
          Fontweight = 'bold')

plt.box(False)

plt.show()
```

```python
#plotting time series for barley
fig, ax = plt.subplots()

ax.plot(df_barley_prices.index, 
        'Price',
        data = df_barley_prices)

locx = plticker.MultipleLocator(base=20.0)

locy = plticker.MultipleLocator(base=20.0)
ax.xaxis.set_major_locator(locx)

ax.yaxis.set_major_locator(locy)

plt.xticks(rotation = 45)

plt.title('Time Series of Barley Prices', 
          fontsize = 18,
          Fontweight = 'bold')

plt.box(False)

plt.show()
```

```python
#plotting time series for sunflower oil
plt.plot('Date', 
        'Price',
        data = df_sunfloweroil_prices)

plt.xticks(rotation = 45)

plt.title('Time Series of Sunflower Oil Prices', 
          fontsize = 18,
          Fontweight = 'bold')

plt.box(False)

plt.show()
```

Ukraine and Russia war started at the end of February 2022 and the commodities which had a very volatile reaction are sunflower oil and wheat which both had a rapid price increase when the conflict started in February 2022.

Maize and Barley did have an increase as well but not as sharply as sunflower oil and wheat.
